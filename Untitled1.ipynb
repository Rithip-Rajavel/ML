```python
#MLP
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

data = load_iris()
X = data.data
y = data.target

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state = 42)

model = StandardScaler()
X_train = model.fit_transform(X_train)
X_test = model.transform(X_test)

mlp = MLPClassifier(hidden_layer_sizes=(10,5),max_iter = 1000,random_state = 42)
mlp.fit(X_train,y_train)

prediction = mlp.predict(X_test)
accuracy = accuracy_score(y_test,prediction)
print(f"Accuracy:{accuracy}")
```

```python
import pandas as pd
import numpy as np


data = pd.read_csv('data.csv')

concepts = np.array(data[:,:-1].values)

target = np.array(data[:,-1].values)

def train(con, tar):
    for i, val in enumerate(tar):
        if val == 'yes':
            specific_h = con[i].copy()
            break

    for i, val in enumerate(con):
        if tar[i] == 'yes':
            for x in range(len(specific_h)):
                if val[x] != specific_h[x]:
                    specific_h[x] = '?'
                else:
                    pass
    return specific_h

print(train(concepts, target))
```

```python
import tensorflow as tf
import numpy as np

# Given initial weights and biases
weights_input_hidden = np.array([[0.1, 0.4], [0.2, 0.5]])
weights_hidden_output = np.array([[0.6, 0.3], [0.7, 0.4]])

biases_hidden = np.array([0.3, 0.4])
biases_output = np.array([0.5, 0.6])

# Training input and output
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

# Convert NumPy arrays to TensorFlow tensors
weights_input_hidden_tf = tf.Variable(weights_input_hidden, dtype=tf.float32)
weights_hidden_output_tf = tf.Variable(weights_hidden_output, dtype=tf.float32)

biases_hidden_tf = tf.Variable(biases_hidden, dtype=tf.float32)
biases_output_tf = tf.Variable(biases_output, dtype=tf.float32)

X_tf = tf.constant(X, dtype=tf.float32)
y_tf = tf.constant(y, dtype=tf.float32)

# Activation function (sigmoid)
def sigmoid(x):
    return 1 / (1 + tf.exp(-x))

# Forward Pass
with tf.GradientTape() as tape:
    hidden_layer_input = tf.matmul(X_tf, weights_input_hidden_tf) + biases_hidden_tf
    hidden_layer_output = sigmoid(hidden_layer_input)

    output_layer_input = tf.matmul(hidden_layer_output, weights_hidden_output_tf) + biases_output_tf
    output_layer_output = sigmoid(output_layer_input)

    # Calculate error
    loss = tf.reduce_sum(tf.square(y_tf - output_layer_output))

# Backward Pass (Automatic Differentiation)
gradients = tape.gradient(loss, [weights_input_hidden_tf, weights_hidden_output_tf, biases_hidden_tf, biases_output_tf])

# Update weights and biases
learning_rate = 0.5
weights_input_hidden_tf.assign_sub(learning_rate * gradients[0])
weights_hidden_output_tf.assign_sub(learning_rate * gradients[1])
biases_hidden_tf.assign_sub(learning_rate * gradients[2])
biases_output_tf.assign_sub(learning_rate * gradients[3])

# Calculate total error
total_error = loss.numpy()
print("Total Error after the first round of Backpropagation:", total_error)
```

```python
import numpy as np

# Given initial weights and biases
weights_input_hidden = np.array([[0.1, 0.4], [0.2, 0.5]])
weights_hidden_output = np.array([[0.6, 0.3], [0.7, 0.4]])

biases_hidden = np.array([0.3, 0.4])
biases_output = np.array([0.5, 0.6])

# Training input and output
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

# Activation function (sigmoid)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of sigmoid function
def sigmoid_derivative(x):
    return x * (1 - x)

# Learning rate
learning_rate = 0.5

# Forward Pass and Backward Pass for one iteration
for i in range(len(X)):
    # Forward Pass
    inputs = X[i]

    hidden_layer_input = np.dot(inputs, weights_input_hidden) + biases_hidden
    hidden_layer_output = sigmoid(hidden_layer_input)

    output_layer_input = np.dot(hidden_layer_output, weights_hidden_output) + biases_output
    output_layer_output = sigmoid(output_layer_input)

    # Backward Pass
    output_error = y[i] - output_layer_output
    output_delta = output_error * sigmoid_derivative(output_layer_output)

    hidden_error = output_delta.dot(weights_hidden_output.T)
    hidden_delta = hidden_error * sigmoid_derivative(hidden_layer_output)

    # Weight updates
    weights_hidden_output += hidden_layer_output.reshape(-1, 1) * output_delta * learning_rate
    weights_input_hidden += inputs.reshape(-1, 1) * hidden_delta * learning_rate

    biases_output += output_delta * learning_rate
    biases_hidden += hidden_delta * learning_rate

# Calculate total error
total_error = 0
for i in range(len(X)):
    hidden_layer_output = sigmoid(np.dot(X[i], weights_input_hidden) + biases_hidden)
    predicted_output = sigmoid(np.dot(hidden_layer_output, weights_hidden_output) + biases_output)
    total_error += np.sum((y[i] - predicted_output) ** 2)

print("Total Error after the first round of Backpropagation:", total_error)
```
