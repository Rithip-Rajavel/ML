```python
from sklearn.cluster import KMeans
import numpy as np

data  = np.array([[0.34, -0.2, 1.13, 4.3],[5.1, -12.6, -7.0, 1.9],[-15.7, 0.06, -7.1, 11.2]])
initial_centroid =  np.array([[1.1, 0.2, -3.1, -0.4], [9.3, 6.1, -4.7,
0.18]])

model = KMeans(n_clusters = 2,init=initial_centroid,n_init=1)

model.fit(data)

final_centoid = model.cluster_centers_
datapoints = model.labels_

print(f"final centroid: {final_centoid}")
for i , labels in enumerate(datapoints):
  print(f"data point{data[i]} and cluster{labels}")

import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
for i ,labels in enumerate(datapoints):
  if labels==0:
    plt.scatter(data[i,0],data[i,1],color='green',marker='o',label='cluster0')
  else:
     plt.scatter(data[i,0],data[i,1],color='blue',marker='s',label='cluster1')
plt.scatter(final_centoid[:, 0],final_centoid[:, 1],color='black',marker='*',label='centroid')
plt.title('Kmeans cluster')
plt.xlabel('feature1')
plt.ylabel('feature2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

data = pd.read_csv('/content/drive/MyDrive/data4_19.csv')

attributes = data.iloc[:,:4].values
y = data.iloc[:,4].values

k = 3
initial_mean_indices = np.random.choice(len(attributes),k,replace=False)
initial_mean_indices
initial_means = attributes[initial_mean_indices]

model = KMeans(n_clusters = k,init= initial_means,n_init=1,max_iter=1000,random_state=42)
model.fit(attributes)

final_centeroid = model.cluster_centers_
assignment = model.labels_

for i, centroid in enumerate(final_centeroid):
    print(f"Cluster {i + 1} Mean: {centroid}")
attribute_names = data.columns[:4]
combinations = [(0, 1), (0, 2), (0, 3), (1, 2), (1, 3), (2, 3)]

plt.figure(figsize=(18, 12))
for i, (x_index, y_index) in enumerate(combinations, 1):
    plt.subplot(2, 3, i)
    for cluster in range(k):
        cluster_indices = np.where(assignment == cluster)[0]
        plt.scatter(attributes[cluster_indices, x_index], attributes[cluster_indices, y_index], label=f'Cluster {cluster+1}')
    plt.xlabel(attribute_names[x_index])
    plt.ylabel(attribute_names[y_index])
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier

# Load dataset
data = pd.read_csv('/content/drive/MyDrive/Data.csv')

# Remove dollar signs and commas from the 'Income' column
data['Income'] = data['Income'].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Select Age and Income columns
age_income = data[['Age', 'Income']]

# Create MinMaxScaler instance and fit_transform the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(age_income)

# Convert the scaled data back to a DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=['Age', 'Income'])

# Record #10's scaled values
record_10 = [[37, 67000]]  # Placeholder values for Age and Income for record #10
scaled_record_10 = scaler.transform(record_10)

# Create a k-NN classifier with k=9
k = 9
knn = KNeighborsClassifier(n_neighbors=k)

# Fit the k-NN classifier on the scaled data with risk factor labels
knn.fit(scaled_df, data['RiskFactor'])

# Predict the risk factor for record #10
predicted_risk_factor = knn.predict(scaled_record_10)
print(f"Predicted Risk Factor for Record #10 using k=9: {predicted_risk_factor}")
```
